```python
import pandas as pd
from bs4 import BeautifulSoup as soup
import requests
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/107.0.0.0 Safari/537.36'}
url='https://www.glassdoor.co.in/Explore/top-information-technology-companies_IS.4,26_ISEC10013.html'
response=requests.get(url,headers=headers)
```

```python
data=soup(response.text)
```

```python
data
```

```python
div_data=data.findAll('div',{'class':'mt-0 mb-std p-std css-1ax1pfu css-errlgf'})
```

```python
company=[]
rating=[]
reviews=[]
salaries=[]
jobs=[]
for i in div_data:
    title=i.findAll('div',{'class':'d-flex'})[1]
    company.append(title.h2.text)
    star=i.findAll('div',{'class':'d-flex'})[2].text
    rating.append(star)
    rev=i.findAll('div',{'class':'d-flex'})[4].text
    reviews.append(rev)
    sal=i.findAll('div',{'class':'d-flex'})[5].text
    salaries.append(sal)
    job=i.findAll('div',{'class':'d-flex'})[6].text
    jobs.append(job)
```

```python
location=[]
size=[]
indus=[]
for i in div_data:
    loc=i.findAll('span',{'class':'d-block mt-0 css-56kyx5'})[0].text
    location.append(loc)
    s=i.findAll('span',{'class':'d-block mt-0 css-56kyx5'})[1].text
    size.append(s)
    ind=i.findAll('span',{'class':'d-block mt-0 css-56kyx5'})[2].text
    indus.append(ind)
```

```python
dict={'Company':company,'Location':location,'Global Company Size':size,'Industry':indus,'Salaries':salaries,'Jobs':jobs,'Rating':rating,'Reviews':reviews}
```

```python
df=pd.DataFrame(dict)
```

```python
final=pd.DataFrame()
for j in range(1,100):
    headers={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/107.0.0.0 Safari/537.36'}
    url='https://www.glassdoor.co.in/Explore/browse-companies.htm?overall_rating_low=3&page={}&sector=10'.format(j)
    response=requests.get(url,headers=headers)
    data=soup(response.text)
    div_data=data.findAll('div',{'class':'mt-0 mb-std p-std css-1ax1pfu css-errlgf'})
    company=[]
    rating=[]
    reviews=[]
    salaries=[]
    jobs=[]
    for i in div_data:
        title=i.findAll('div',{'class':'d-flex'})[1]
        company.append(title.h2.text)
        star=i.findAll('div',{'class':'d-flex'})[2].text
        rating.append(star)
        rev=i.findAll('div',{'class':'d-flex'})[4].text
        reviews.append(rev)
        sal=i.findAll('div',{'class':'d-flex'})[5].text
        salaries.append(sal)
        job=i.findAll('div',{'class':'d-flex'})[6].text
        jobs.append(job)
        location=[]
    size=[]
    indus=[]
    for i in div_data:
        loc=i.findAll('span',{'class':'d-block mt-0 css-56kyx5'})[0].text
        location.append(loc)
        s=i.findAll('span',{'class':'d-block mt-0 css-56kyx5'})[1].text
        size.append(s)
        try:
            ind=i.findAll('span',{'class':'d-block mt-0 css-56kyx5'})[2].text
            indus.append(ind)
        except ValueError:
            indus.append('None')
    dict={'Company':company,'Location':location,'Global Company Size':size,'Industry':indus,'Salaries':salaries,'Jobs':jobs,'Rating':rating,'Reviews':reviews}
    df=pd.DataFrame.from_dict(dict, orient='index')
    df = df.transpose()
    final=final.append(df, ignore_index=True)
```

```python
final.to_csv('GlassDoor_Data.csv')
```
